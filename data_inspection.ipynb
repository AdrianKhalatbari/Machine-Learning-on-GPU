```python
# Imports
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = np.loadtxt('data/MLoGPU_data3_train.csv', delimiter=',')
X = data[..., :-1]
y = data[..., -1]

print(X.shape, y.shape)
```

```python
# Print classes and number of samples per class
classes = np.unique(y)
num_classes = len(classes)
print(classes)

class_samples = np.zeros(num_classes)

for i in range(len(y)):
    class_samples[int(y[i] - 1)] += 1
    
fig, ax = plt.subplots()
bars = ax.bar(classes, class_samples)
ax.bar_label(bars, label_type='edge')
plt.xticks(range(num_classes+1))
plt.xlabel('Classes')
plt.ylabel('Samples')
plt.show()
```

```python
df = pd.DataFrame(X, columns=[f"f{i}" for i in range(7)])

df["class"] = y

sns.pairplot(
    df,
    hue="class",
    palette="tab10",
    corner=True
)

plt.show()
```
